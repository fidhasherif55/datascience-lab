```python
n = 5 # Number of rows

for i in range(n):
    # Print leading spaces
    print(' ' * (n - i - 1), end='')
    # Print stars
    print('* ' * (i + 1))
```

```python
n = 5 # Number of rows for half of the diamond

# Upper half of the diamond (pyramid)
for i in range(n):
    print(' ' * (n - i - 1), end='')
    print('* ' * (i + 1))

# Lower half of the diamond (inverted pyramid, excluding the middle row already printed)
for i in range(n - 1, 0, -1):
    print(' ' * (n - i), end='')
    print('* ' * i)
```

```python
n = 5 # Number of rows for half of the diamond (determines width/height)

# Upper half of the hollow diamond
for i in range(n):
    # Print leading spaces
    print(' ' * (n - i - 1), end='')

    # Print first star
    print('*', end='')

    # Print inner spaces if not the top row
    if i > 0:
        print(' ' * (2 * i - 1), end='')
        print('*', end='') # Print second star
    print() # Move to the next line

# Lower half of the hollow diamond (inverted, excluding the middle row already printed)
# Loop from n-2 down to 0
for i in range(n - 2, -1, -1):
    # Print leading spaces
    print(' ' * (n - i - 1), end='')

    # Print first star
    print('*', end='')

    # Print inner spaces if not the bottom row
    if i > 0:
        print(' ' * (2 * i - 1), end='')
        print('*', end='') # Print second star
    print() # Move to the next line
```

```python
from math import comb

rows = 6 # Number of rows for Pascal's triangle

for i in range(rows):
    # Print leading spaces for alignment
    print(' ' * (rows - i - 1), end='')
    for j in range(i + 1):
        # Calculate and print the binomial coefficient C(i, j)
        print(comb(i, j), end=' ')
    print()
```

```python
n = 5 # Number of rows for half of the butterfly

# Upper half of the butterfly
for i in range(1, n + 1):
    print('*' * i, end='')
    print(' ' * (2 * (n - i)), end='')
    print('*' * i)

# Lower half of the butterfly
for i in range(n, 0, -1):
    print('*' * i, end='')
    print(' ' * (2 * (n - i)), end='')
    print('*' * i)
```

```python
print("\n--- Solid Diamond Pattern ---\n")

n = 5 # Number of rows for half of the diamond

# Upper half of the diamond (pyramid)
for i in range(n):
    print(' ' * (n - i - 1), end='')
    print('* ' * (i + 1))

# Lower half of the diamond (inverted pyramid, excluding the middle row already printed)
for i in range(n - 1, 0, -1):
    print(' ' * (n - i), end='')
    print('* ' * i)


print("\n--- Butterfly Pattern ---\n")

n = 5 # Number of rows for half of the butterfly

# Upper half of the butterfly
for i in range(1, n + 1):
    print('*' * i, end='')
    print(' ' * (2 * (n - i)), end='')
    print('*' * i)

# Lower half of the butterfly
for i in range(n, 0, -1):
    print('*' * i, end='')
    print(' ' * (2 * (n - i)), end='')
    print('*' * i)
```
